## 3 Data Wrangling

### 3.1 Dealing with missing values

```python
#Identify missing values
shark_merged.isna().sum().sort_values(ascending = False)
```

```python
#Store missing values in a variable
missing_values = shark_merged_copy.isna().sum().sort_values(ascending = False)
```

```python
#Find out the length of the df
len(shark_merged_copy)
```

```python
#Determine the percentage of missing values
percentage_missing_values = round(missing_values/len(shark_merged_copy),2)
percentage_missing_values
```

```python
#Analyze column entrepreneurs
shark_merged_copy[shark_merged_copy["entrepreneurs"].isna()]["season"].value_counts()
```

```python
#Look at similar columns, and start with column "season"
shark_merged_copy["season"].value_counts()
```

```python
#See seasons and episodes (first letter refers to season)
print(sorted(set(shark_merged_copy["Season_Epi_code"])))

```

```python
#Check merged copy with episodes equal to or higher than 701
shark_merged_copy.loc[shark_merged_copy["Season_Epi_code"] >= 701]
```

```python
#Drop rows of episodes equal to or higher than 701
shark_merged_copy.drop(shark_merged_copy.loc[shark_merged_copy["Season_Epi_code"] >= 701].index, inplace = True)
```

```python
#Check if the numbers have been removed
print(sorted(set(shark_merged_copy["Season_Epi_code"])))
```

```python
#See which values are still missing
shark_merged_copy.isna().sum()
```

```python
shark_merged_copy.dtypes
```

```python
shark_merged_copy['episode'] = shark_merged_copy['episode'].astype(int, errors='ignore')
shark_merged_copy
```

```python
#Check if the missing values in column "episode season" are related to "deal shark". Closed deals by DJ are overrepresented.
shark_merged_copy[shark_merged_copy["episode_season"].isna()]["Deal_Shark"].value_counts()
```

```python
#Check if the missing values in column "episode_season" are related to "Deal_Status". No relation could be established due to an even distribution of deals closed and not closed.
shark_merged_copy[shark_merged_copy["episode_season"].isna()]["Deal_Status"].value_counts()
```

```python
#Check relation of missing values of "episode season" (from shark_ent) and "Season_Epi_code" (from shark_inv). Some episodes are listed multiple times.
shark_merged_copy[shark_merged_copy["episode_season"].isna()]["Season_Epi_code"].value_counts()
```

```python
#Drop duplicate rows in column "Season_Epi_code"

shark_merged_copy["Season_Epi_code"].drop_duplicates(keep = "first", inplace = True)
```
